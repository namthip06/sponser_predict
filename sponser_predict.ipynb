```python
df = pd.read_csv('/content/drive/MyDrive/Work/data_sponser.csv')
df
```

```python
from pythainlp.corpus.common import thai_stopwords
from pythainlp import word_tokenize
import re

stopwords = list(thai_stopwords())

def text_process(text):
    final = re.sub(r'#\S+', ' ', text)
    final = "".join(u for u in final if re.match(r'[a-zA-Z‡∏Å-‡πô]', u) and u != '‡πÜ')
    final = word_tokenize(final)
    final = " ".join(word for word in final)
    final = " ".join(word for word in final.split() if word.lower not in stopwords)
    return final

def count_hashtags(text):
    parts = text.split('#')
    hashtags = [part.split()[0] for part in parts[1:]]
    return len(hashtags)
```

```python
df['text_tokens'] = df['text'].apply(text_process)
df['hashtags'] = df['text'].apply(count_hashtags)
#df['spon'] = df['category'].apply(lambda x: 1 if x=='nospon' else 0)
df
```

# **Sklearn**
Logistic Regression + bag of word

**version** hashtag

```python
def hashtag_process(text, num):
    final = text + " #" * num
    return final

df['text_hashtags'] = df.apply(lambda row: hashtag_process(row['text_tokens'], row['hashtags']), axis=1)
df['text_hashtags']
```

```python
from sklearn.model_selection import train_test_split

X = df['text_hashtags']
y = df['category']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cvec = CountVectorizer(analyzer=lambda x:x.split(' '))
cvec.fit_transform(X_train)
```

```python
train_bow = cvec.transform(X_train)
pd.DataFrame(train_bow.toarray(), columns=cvec.get_feature_names_out())
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_bow, y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

test_bow = cvec.transform(X_test)
test_predictions = lr.predict(test_bow)
print(classification_report(test_predictions, y_test))
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
X_tf = df[['text_tokens']]
y_tf = df['spon']
X_tf_train, X_tf_test, y_tf_train, y_tf_test = train_test_split(X_tf, y_tf, test_size=0.2, random_state=1)

X_train.head()
```

```python
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

```python
def get_sentence_embeding(sentences):
    preprocessed_text = bert_preprocess(sentences)
    return bert_encoder(preprocessed_text)['pooled_output']
```

```python
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)

l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs['pooled_output'])
l = tf.keras.layers.Dense(1, activation='sigmoid', name="output")(l)

model = tf.keras.Model(inputs=[text_input], outputs = [l])
```

```python
model.summary()
```

```python
METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall')
]

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=METRICS)
```

```python
model.fit(X_tf_train, y_tf_train, epochs=10)
```

```python
model.evaluate(X_tf_train, y_tf_train)
```

```python
y_predicted = model.predict(X_tf_test)
y_predicted = y_predicted.flatten()
y_predicted = np.where(y_predicted > 0.5, 1, 0)
y_predicted
```

```python
print(classification_report(y_tf_test, y_predicted))
```
